## groupby()

```python
weather_data = {
    'day': ['1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017', '1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017',
            '1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017'],
    'city': ['mumbai', 'mumbai', 'mumbai', 'mumbai', 'newyork', 'newyork', 'newyork', 'newyork',
             'paris', 'paris', 'paris','paris'],
    'temperature' : [32, 36, 28, 33, 90, 85, 87, 92, 45, 50, 54, 42],
    'windspeed' : [6,7,12,7,5,12,15,5,20,13,8,10],
    'event' : ['rain', 'sunny', 'snow', 'sunny', 'sunny', 'fog', 'fog', 'rain', 'sunny', 'cloudy', 'cloudy', 'cloudy']    
}

wdf = pd.DataFrame(weather_data)
wdf
```

```python
g = wdf.groupby('city')
g
```

```python
for city, city_df in g:
    print(city)
    print(city_df)
```

```python
g.get_group('paris')
```

```python
g = wdf.groupby('event')
g
```

```python
for event, event_df in g:
    print(event)
    print(event_df)
```

```python
g.get_group('rain')
```

```python
g.max()
```

```python
g.std()
```

```python
g.describe()
```

```python
%matplotlib inline
g.plot()
```
