```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## lineplot()

```python
data_frame = {
    'height': np.random.randint(25, 50, 25),
    'weight': np.random.randint(50, 100, 25)
}

df = pd.DataFrame(data_frame)
df
```

```python
sns.lineplot(x = 'height', y = 'weight', data = df)
plt.show()
```

```python
df.plot(kind = 'scatter', x = 'height', y = 'weight')
plt.show()
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
tips_df.to_csv('tips.csv')
```

```python
sns.lineplot(x = 'total_bill', y = 'tip', data = tips_df)
plt.show()
```

```python
sns.lineplot(x = 'size', y = 'total_bill', data = tips_df, hue = 'day')
plt.show()
```

## histogram / distplot() 

```python
bins = np.arange(0, 65, 5)
print(bins)
plt.figure(figsize = (16,9))
sns.set()
sns.distplot(tips_df['total_bill'], bins, hist_kws = {'edgecolor': 'black', 'rwidth': 0.90})
plt.xticks(bins)
plt.yticks(np.arange(0, 0.06, 0.002))
plt.show()
```

## scatterplot()

```python
titanic_df = sns.load_dataset("titanic")
titanic_df
```

```python
titanic_df.to_csv('titanic.csv')
```

```python
plt.figure(figsize = (16, 9))
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue = 'sex')
```

```python
plt.figure(figsize = (16, 9))
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue = 'sex', style = 'who')
```

```python
plt.figure(figsize = (16, 9))
sns.scatterplot(x = 'who', y = 'fare', data = titanic_df, hue = 'alive', style = 'alive')
```

```python
plt.figure(figsize = (16, 9))
sns.scatterplot(x = 'class', y = 'alive', hue = 'sex', data = titanic_df)
```
